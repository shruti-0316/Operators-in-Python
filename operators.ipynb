**1.Arithmetic Operators in Python.**

```python
a = 10
b = 3
print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}")
print(f"a % b = {a % b}")
print(f"a ** b = {a ** b}")
print(f"a // b = {a // b}")
```

**2.Assignment Operators in Python.**

```python
a = 10
b = 3
a += b
print(f"a += b => {a}")
a -= b
print(f"a -= b => {a}")
a *= b
print(f"a *= b => {a}")
a /= b
print(f"a /= b => {a}")
a %= b
print(f"a %= b => {a}")
a **= b
print(f"a **= b => {a}")
a //= b
print(f"a //= b => {a}")
```

**3.Comparison Operators in Python.**

```python
a = 10
b = 3
a += b
print(f"a += b => {a}")
a -= b
print(f"a -= b => {a}")
a *= b
print(f"a *= b => {a}")
a /= b
print(f"a /= b => {a}")
a %= b
print(f"a %= b => {a}")
a **= b
print(f"a **= b => {a}")
a //= b
print(f"a //= b => {a}")
```

**4.Logical Operators in Python.**

```python

a = 10
b = 3
c = 20
print(a < b and a > c)
print(a < b or a > c)
print(not a > b)
```

**5. Identity Operators in Python.**

```python

a = 10
b = 3
print(a is b)
print(a is not b)
```

**6.Membership Operators in Python.**

```python

list = [1, 4, 10, 40, 5]
a = 10
b = 3
print(a in list)
print(b in list)
print(a not in list)
print(b not in list)
print('s' in 'btechsmartclass')
```
